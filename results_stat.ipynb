```python
import pandas as pd

file_paths = [
    'experiment3/final/FINAL_FINAL_ALL.csv',
    'experiment2/final/FINAL_FINAL_ALL.csv',
    'experiment/final/FINAL_FINAL_ALL.csv'
]

dataframes = [pd.read_csv(file, index_col=0).round(4) for file in file_paths]
stacked = pd.concat(dataframes, keys=range(len(dataframes)))

columns_to_keep = ['KL_median', 'Chi-Square_median', 'Column Pair Trend']


mean_df = stacked.groupby(level=1).mean()
std_df = stacked.groupby(level=1).std()

mean_df = mean_df[columns_to_keep]
std_df = std_df[columns_to_keep]

result = mean_df.round(4).astype(str) + " ± " + std_df.round(4).astype(str)

result.to_csv('FINAL_final_results_mean_std.csv')
print("Results saved to 'final_results_mean_std.csv'")
result
```
